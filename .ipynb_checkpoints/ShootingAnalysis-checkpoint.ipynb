# Inspecting states

```python
plt.figure(figsize = (10,10))
plt.pie(x = df['state'].value_counts(),labels = df['state'].unique());
```

```python
m = folium.Map(location = [37,-102], zoom_start = 4)
```

```python
states = df['state'].value_counts()
```

```python
folium.Choropleth(
    geo_data='https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/us-states.json',
    name='choropleth',
    data=states,
    key_on='feature.id',
    fill_color='BuPu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Cases per state'
).add_to(m);
```

```python
m
```

```python
plt.figure(figsize = (38,12))
plt.subplot(2,2,1)
sns.countplot(x = df['state'],hue=df['body_camera'])
plt.title('cases using cameras per state')
plt.show();
```

_We have a lot of cases registered in California with and without camera usage. In order to see which state uses more camera, we need to take a look at the proportions_

```python
dic = {}
for state in df['state'].unique():
    tam = len(df[df['state'] == state])
    notusing = df[(df['state'] == state) & (df['body_camera'] == False)]
    prop = (len(notusing)*100)/tam
    dic[state] = prop
```

```python
stateproportion = pd.DataFrame(list(dic.items()),columns = ['state','proportion']) 
```

```python
sns.hist(stateproportion)
```
