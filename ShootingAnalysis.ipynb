# Importing useful libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
df = pd.read_csv('fatal-police-shootings-data.csv')
df['date'] =  pd.to_datetime(df['date'], format='%Y-%m-%d')
df.columns
```

```python
df.shape
```

_5416 rows and 14 columns_

**Checking missing values**

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['gender'])
```

_We have only 2 missing values over the gender column so i'll drop them_


```python
df.head(3)
```

```python
cardinality ={}
for col in df.columns:
    cardinality[col] = df[col].nunique()
```

```python
cardinality
```

```python
df['name'].duplicated().sum()
```

```python
df.duplicated().sum()
```

_It seems that we have some duplicated names but not duplicated cases, which is weird. So im just going to drop the ones w duplicated names and consider the first one that appeared_

```python
df = df.drop_duplicates(subset=['name'])
```

# Age inspecting

```python
sns.boxplot(df['age'],orient='v').set_title('Distribution of cases over ages')
```

```python
sns.distplot(df['age'].dropna(),kde = False)
```

_Looking at the shape of the histogram and the dot values at the boxplot, we can tell it has a right skewed distribution over the ages and some outliers around the 80's_

```python
print('Mean age: {}'.format(df['age'].dropna().mean()))
print('Median: {}'.format(pd.Series.median(df['age'].dropna())))
print('Moda: {}'.format(df['age'].mode(dropna = True).values))
```

#### Age and gender distribution 

```python
sns.barplot(x = df['gender'], y = df['age'])
```

_Our age distribution is similar to both genders apparently_

```python
sns.countplot(df['gender'],palette='dark')
```

_Our dataset is mainly formed by males who have been shot_

```python
print('Percentage distribution over gender:\nMale: {}%\nFemale: {}%\n'.format(len(df[df['gender'] == 'M'])*100/len(df),
                                                                               len(df[df['gender'] == 'F'])*100/len(df)))
```

```python
sns.countplot(df['manner_of_death'],hue = df['gender'],palette='bright')
```

```python
print('Woman that got shot and Tasered count: {}' .format(len(df[(df['gender']=='F' ) & (df['manner_of_death'] == 'shot and Tasered')])))
```
